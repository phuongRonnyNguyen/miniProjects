```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

pwd = os.getcwd()
```

```python
data = pd.read_csv(pwd + '/coffee_exports.csv', index_col = 0)
data_modified = data.sort_values('all_years', ascending=False).head(11).copy()
```

```python
data_modified = data_modified.drop(data_modified.index[0])
data_modified = data_modified.drop(columns=['all_years'])
data_modified = data_modified.T
data_modified.index.names = ['Years']
data_modified.columns.names = ['Countries']
```

```python
data_modified.columns
```

```python
data_year = data_modified.index
data_VietNam = data_modified[['Brazil', 'Viet Nam', 'Colombia', 'Indonesia', 'India', 'Guatemala',
       'Honduras', 'Uganda', 'Mexico', 'Peru']]
```

```python
figure = plt.figure(figsize=(20,10))
plt.plot(data_year, data_VietNam)
plt.title('Vietnam Coffee Export')
plt.show()
```
