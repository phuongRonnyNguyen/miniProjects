```python
def GetData(file_name = str, df_name = str):
    if '.csv' in file_name:
        globals()[df_name] = pd.read_csv(f'{pwd}/{file_name}')
        return 
    if '.excel' in file_name:
        globals()[df_name] = pd.read_excel(f'{pwd}/{file_name}')
        return
    if '.json' in file_name:
        globals()[df_name] = pd.read_json(f'{pwd}/{file_name}')
        return
        
```
